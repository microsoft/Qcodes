```python
import matplotlib.pyplot as plt

import qcodes
from qcodes.instrument_drivers.rigol.DS4034 import Rigol_DS4035
```

```python
# You can find out the address of the instrument by pressing the "menu" button on the scope and going to IP settings.  
scope = Rigol_DS4035("rigol_scope", "TCPIP::169.254.180.17::INSTR")
```

```python
# We have connected the 10 MHz output of the scope on channel 1
data = scope.get_wave_form(1000, 1)
```

```python
fig, ax = plt.subplots()
t, voltage = data
ax.plot(t * 1E6, voltage)
ax.set_xlabel("time [ms]")
ax.set_ylabel("voltage [V]")
plt.show()
```

```python
# Lets see how we can access the individual channels 
amplitude = scope.channels[0].amplitude()
```

```python
print("the amplitude is {:.3} V".format(amplitude))
```
